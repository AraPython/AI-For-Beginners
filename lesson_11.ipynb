```python
import requests
from bs4 import BeautifulSoup

URL = "https://www.maannews.net/news"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
print(type(soup))
print(soup)
```

# **Find Elements by ID**
In an HTML web page, every element can have an id attribute assigned. As the name already suggests, that id attribute makes the element uniquely identifiable on the page. You can begin to parse your page by selecting a specific element by its ID.

```python
results = soup.find(id="banner-b-75")
print(results)
```

```python
import json
results = soup.find(class_="row is-multiline")
#print(results)
elements = results.find_all("div", class_="column is-4")
#print(elements)
for element in elements:
    url = str(element).split("\"")[5]
    date = element.find("div",class_="list-6__date")
    title = element.find("div",class_="list-6__title")
    #image = str(element.find( "div",class_="list-6__image")).split("\"")[7].split("?")[0]
    image = str(element.find( "div",class_="list-6__image")).split("\"")[7].split("?")[0]
    print(f"date.text = {date.text}")
    print(f"title.text = {title.text}")
    print(f"url = {url}")
    print(f"image url = {image}")
    print()
```
