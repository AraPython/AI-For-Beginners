1- Creating and Using a Class 


```python
class Dog:
  """A simple attempt to model a dog."""
  def __init__(self, name, age, color= "Black"):
  #def __init__(self):
    """Initialize name and age attributes."""
    self.name = name
    self.age = age
    self.color = color
  def sit(self,h="Haow Haow"):
    """Simulate a dog sitting in response to a command."""
    print(f"{self.name} is now sitting.")
    print(h)
  def roll_over(self):
    """Simulate rolling over in response to a command."""
    print(f"{self.name} rolled over!")
```

Creating one instance

```python
my_dog = Dog('Rex', 2)
print(f"My dog's name is {my_dog.name}.")
print(f"My dog is {my_dog.age} years old.")
print(f"My dog color is {my_dog.color}.")

my_dog = Dog('Max', 5, color="Blue")
print(f"My dog's name is {my_dog.name}.")
print(f"My dog is {my_dog.age} years old.")
print(f"My dog color is {my_dog.color}.")
```

Creating two instances

```python
dog1 = Dog('Willie', 6)
dog2 = Dog('Lucy', 3)
print(f"My dog's name is {dog1.name}.")
print(f"My dog is {dog1.age} years old.")
dog1.sit(h="Hoooo")




print(f"\nYour dog's name is {dog2.name}.")
print(f"Your dog is {dog2.age} years old.")
dog2.sit()
```

```python
class Car:

  """A simple attempt to represent a car."""
  def __init__(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year
    self.odometer_reading = 0
    self.color = "color"


  def get_descriptive_name(self):
    #long_name = f"{self.year} {self.make} {self.model}"
    long_name = str(self.year) + " " + self.make + " " + self.model + "Color is " + self.color
    """
    do somthing
    """
    return long_name.title()

    
  def read_odometer(self):
    #print("This car has "+self.odometer_reading+" miles on it.")
    print(f"This car has {self.odometer_reading} miles on it.")

  def get_Color(self):
    #print(f"This car color id {self.color}")
    return self.color

  def set_Color(self, color="white"):
    self.color = color  
  
  def update_odometer(self, mileage):
    if mileage >= self.odometer_reading:
      self.odometer_reading = mileage
    else:
      print("You can't roll back an odometer!")


  def increment_odometer(self, miles):
    self.odometer_reading += miles

class ElectricCar(Car):
  """Represent aspects of a car, specific to electric vehicles."""
  def __init__(self, make, model, year , color ):
    """Initialize attributes of the parent class."""
    super().__init__(make, model, year , color )
    
```

```python
car1 = Car(make="Honda",model="civic",year=2020 )
print(car1.get_descriptive_name())
print(car1.read_odometer())
car1.update_odometer(50)
car1.update_odometer(40)

#car1.odometer_reading=40
print(car1.read_odometer())
```

```python
car2 = Car(make="Honda",model="civic",year=2020, color="Blue" )
print(car2.get_Color())

car2.set_Color(color="Yellow")
print("new color is ") 
car2.get_Color()
```
